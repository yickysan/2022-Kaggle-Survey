## What Are The Most Popular Platforms For Studying Data Skills

To deal with the multiple choice questions, I've created a function `summarise_question`. This function takes in the question number as an argument e.g to summrise question 7 we pass the argument `summarise_question("Q7")`. All of the columns in the resulting dataframe for the questions will be converted to rows using the `pivot_longer()` function from `tidyr` and then will be grouped by the response to the question and aggregated by the count. This makes it easy to see how many times respondents picked a particular option in the multiple choice questions.

To answer the question on which learning platform is the most popular, `"Q6" and "Q7"` will be passed as an argument for our function. A supplementary data was provided by kaggle on answer choices. You can read about the answer choices <a href = "">here</a>

```r
# function to summarise mutiple choice questions
summarise_question <- function(Q){
question <- survey_2022 %>% select(starts_with(Q)) # select columns that starts with the question number
colnames(question) <- question %>% head(1) # renames the columns with the first row of the data
question <- question[-1,,] # selects every row from the second row

# convert the data frame to long format, group and summarise by count
question <- question %>% pivot_longer(
cols = everything(),
names_to = "cols",
values_to = "value") %>% drop_na() %>% group_by(value) %>% summarise(count = n()) %>% arrange(-count)

return(question)
    
}
```

```r
# function to control plot size
fig_size <- function(width, height){
options(repr.plot.width = width, repr.plot.height = height)
}
```

## How Much Do Data Professionals Earn?
People don't just go into data roles because they are passionate about data, sometimes money can be the biggest motivator. There's always talk online about the average salary of a data scientist being just over 100k dollars and people see this and are pushed to start a career in data. To answer the question on how much data professionals earn, the required columns are `Q4, Q23, Q29`.

```r
compensation <- survey_2022[-1,,] %>% select(c(Q4, Q23, Q29)) %>% drop_na()
```

```r
compensation %>% head()
```

The column `Q29` contains data on the salaries and it is a character(string) data type.  It is also a range of values. The salary is going to be gotten by converting the values from character(string) to numeric(float) and taking the averag e of th range.

```r
compensation %>% pull(Q29) %>% unique()
```

```r
compensation <- compensation %>% mutate(
    Q29 = map_chr(Q29, function(x) str_replace(x, "\\$", "")),
    Q29 = map_chr(Q29, function(x) str_replace_all(x, ",", "")),
    Q29 = map_chr(Q29, function(x) str_replace(x, ">", ""))
              ) %>% mutate(
    Q29 = str_split(Q29, "-")) %>% mutate( 
        # if the first value in the range is 0, it is replaced with the second value so the value stays the same when taking average
        lower_bound = unlist(map(Q29, function(x) if_else(x[1] == "0", x[2], x[1]))),
        
        # if the first value is 1000000, the second value is replaced with the first value to avoid NA
        upper_bound = unlist(map(Q29, function(x) if_else(x[1] == "1000000", x[1], x[2]))) 
                                            ) %>% mutate(
        lower_bound = as.numeric(lower_bound),
        upper_bound = as.numeric(upper_bound)
        ) %>% mutate(compensation = round((lower_bound + upper_bound)/2, -3)) # taking the average between the salary range and rounding to the nearest thousand
```

```r
fig_size(18, 10)
filt2 <- str_detect(compensation$Q23, pattern)
data_compensation <- compensation %>% filter(filt2)
data_compensation %>% ggplot(aes(x=compensation, y=Q23)) + 
geom_boxplot() +
scale_x_continuous(breaks = seq(0, 1000000, 100000), labels = scales::label_comma()) +
labs(
    title = "Distribution Of Salary For Different Data Roles",
    x = "Salary($)",
    y = "Data Role"
    ) +
theme_bw() +
theme(
plot.title = element_text(size = 25, face = "bold"),
axis.title.x = element_text(size = 12, face = "bold"),
axis.text = element_text(size = 12, face = "bold")
    )
```

The boxplot shows the distribution of the salaries for the different data roles. The square box represents the interquartile range. The bottom and top of the box represents the 25th and 75th percentile respectively while the line dividing the box in two is the median salary. The line beneath and above the boxes represents the min and max salaries respectively while the dots represents outliers. From the boxplot, Data Administrators have the lowest median salary, while Data Architects have the highest median salary. To look into the salary more ganularly, the salaries will be broken down by region also.

For the sake of simplicity, countries will be grouped into the following regions:
* Africa
* Australia
* Asia
* Europe
* North America
* South America

```r
for (value in data_compensation %>% pull(Q4) %>% unique()){print(value)} # printing every country name in the data
```

```r
europe <- c("France", "Germany", "Italy", "Turkey", "Spain",
            "United Kingdom of Great Britain and Northern Ireland", "Netherlands",
            "Portugal", "Russia", "Poland", "Czech Republic", "Romania", "Ukraine", "Belgium", "Ireland")

asia <- c("Singapore", "Taiwan", "Japan", "Indonesia", "India", "Pakistan",
          "Philippines", "South Korea", "Thailand", "Iran, Islamic Republic of...", 
          "Viet Nam", "Saudi Arabia", "Bangladesh", "Israel", "China", "Sri Lanka", 
          "Hong Kong (S.A.R.)", "United Arab Emirates","Nepal", "Malaysia")

africa <- c("Cameroon", "Nigeria", "Ethiopia", "Tunisia", "Egypt", "Ghana", 
            "South Africa","Kenya", "Morocco", "Algeria", "Zimbabwe")

north_america <- c("United States of America", "Canada", "Mexico")

south_america <- c("Brazil", "Argentina", "Chile", "Colombia",
                    "Peru", "Ecuador")
```

```r
data_compensation <- data_compensation %>% mutate(
    region = case_when(
                Q4 %in% europe ~ "Europe",
                Q4 %in% africa ~ "Africa",
                Q4 %in% asia ~ "Asia",
                Q4 %in% north_america ~ "North America",
                Q4 %in% south_america ~ "South America",
                Q4 == "Australia" ~ "Australia",
                Q4 == "I do not wish to disclose my location" | Q4 == "Other" ~ "Unknown"
                     )
    )
```

```r
data_compensation %>% filter(!region == "Unknown") %>% ggplot(aes(x=compensation, y=Q23, colour = region)) + 
geom_boxplot() +
scale_x_continuous(breaks = seq(0, 1000000, 100000), labels = scales::label_comma()) +
labs(
    title = "Distribution Of Salary For Different Data Roles By Region",
    x = "Salary($)",
    y = "Data Role"
    ) +
theme_bw() +
theme(
plot.title = element_text(size = 25, face = "bold"),
axis.title.x = element_text(size = 12, face = "bold"),
axis.text = element_text(size = 12, face = "bold")
    )
```

The boxplot shows the distribution of salaries for the different data roles in different regions. When we look at it more granularly like this, it shows that machine learning engineers in North America earn the most money while data analyst in africa earn the least. To summarise the salary for the different data roles in different regions, I'm going to be taking the median salary instead of the mean as the mean is susceptible to outliers.

```r
data_compensation_summary <- data_compensation %>% group_by(Q23, region) %>% summarise(
    median_salary = compensation %>% median(),
    mean_salary = compensation %>% mean(),
    .groups = "drop") %>% arrange(-median_salary) %>% filter(!region == "Unknown")

data_compensation_summary
```

```r
fig_size(25, 15)
data_compensation_summary %>% ggplot(
    aes(x=region, y=median_salary, fill=region)) +
geom_bar(stat = "identity", show.legend = F) + 
geom_text(aes(label = median_salary), vjust = -0.2, colour = "blue" ) +
labs(
    title = "Median salary For Different Data Roles By Region",
    y = "Median Salary($)") + 
scale_fill_brewer(palette = 9) +
facet_wrap(~Q23, scales = "free_y") +
theme(plot.title = element_text(size = 25, face = "bold"),
axis.title.y = element_blank(),
axis.title.x = element_text(size = 12, face = "bold"),
axis.ticks = element_blank(),
axis.text.y = element_blank(),
axis.text.x = element_text(size = 12, face = "bold"),
panel.background = element_blank(),
panel.grid.major.y = element_line(colour = "gray", linewidth = 0.2),
strip.background = element_blank(),
strip.text = element_text(size = 15, face = "bold"))
```
